```python
import csv
from datetime import datetime

# Base path for input and output files, adjust this to your actual paths
base_input_path = '/Users/oscarli/Desktop/GitHub/eric-trial-data/fert_man_file_sys{}.csv'
base_output_path = '/Users/oscarli/Desktop/GitHub/eric-trial-data/fer_{}_sys{}.txt'

# Function to parse the date and extract the year
def get_year_from_date(date_str):
    try:
        return datetime.strptime(date_str, "%d%m%Y").year
    except ValueError:
        return datetime.strptime(date_str, "%d%m%y").year

# Loop through each system
for sys_num in range(1, 6):  # From sys1 to sys5
    yearly_data = {}
    
    # Adjust file paths for the current system
    file_path = base_input_path.format(sys_num)
    
    # Read the CSV file for the current system
    with open(file_path, mode='r') as file:
        reader = csv.reader(file)
        next(reader)  # Skip the header line
        for row in reader:
            date_str = row[0]
            year = get_year_from_date(date_str)
            if year not in yearly_data:
                yearly_data[year] = []
            yearly_data[year].append(','.join(row))
    
    # Write each year's data to a separate text file for the current system
    for year, lines in yearly_data.items():
        output_file_path = base_output_path.format(year, sys_num)
        with open(output_file_path, 'w') as file:
            for line in lines:
                file.write(line + '\n')
        print(f'Data for {year} written to {output_file_path}')
```

```python
import csv
import os
from datetime import datetime

# Base path for input files, adjust this to your actual paths
base_input_path = '/Users/oscarli/Desktop/GitHub/eric-trial-data/fert_man_file_sys{}.csv'
# Base output directory, adjust to where you want to store the output folders
base_output_dir = '/Users/oscarli/Desktop/GitHub/eric-trial-data/'

# Function to parse the date and extract the year
def get_year_from_date(date_str):
    try:
        return datetime.strptime(date_str, "%d%m%Y").year
    except ValueError:
        # Handle dates with missing leading zeros
        return datetime.strptime(date_str, "%d%m%y").year

# Loop through each system
for sys_num in range(1, 6):  # From sys1 to sys5
    yearly_data = {}
    
    # Adjust file path for the current system
    file_path = base_input_path.format(sys_num)
    
    # Create a directory for the current system if it doesn't exist
    system_output_dir = os.path.join(base_output_dir, f'fert_sys_{sys_num}')
    os.makedirs(system_output_dir, exist_ok=True)
    
    # Read the CSV file for the current system
    with open(file_path, mode='r') as file:
        reader = csv.reader(file)
        next(reader)  # Skip the header line
        for row in reader:
            date_str = row[0]
            year = get_year_from_date(date_str)
            if year not in yearly_data:
                yearly_data[year] = []
            yearly_data[year].append(','.join(row))
    
    # Write each year's data to a separate text file within the system-specific folder
    for year, lines in yearly_data.items():
        output_file_path = os.path.join(system_output_dir, f'fert_{year}_sys{sys_num}.txt')
        with open(output_file_path, 'w') as file:
            for line in lines:
                file.write(line + '\n')
        print(f'Data for {year} written to {output_file_path}')
```
