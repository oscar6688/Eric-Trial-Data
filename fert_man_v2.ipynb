## 5/14/24 liquid fertilizer changes

```python
import pandas as pd

# Function to add a leading zero to dates with length 7
def add_leading_zero(date_str):
    return '0' + date_str if len(date_str) == 7 else date_str

# Function to perform calculations if 'fertilizer type' contains 'liquid fertilizer'
def calculate_fertilizer(row):
    if 'liquid fertilizer' in row['fertilizer type']:
        row['fert NH4-N (g m-2).1'] = round(row['fert manure N (g m-2) adj'] * 0.13, 3)
        row['fert NO3-N banded (g m-2)'] = round(row['fert manure N (g m-2) adj'] * 0.09, 3)
        row['fert manure N (g m-2) adj'] = round(row['fert manure N (g m-2) adj'] - (row['fert NH4-N (g m-2).1'] + row['fert NO3-N banded (g m-2)']), 3)
    return row

# List of file paths
file_paths = [
    'fert_man_data/fert_man_file_sys1.csv',
    'fert_man_data/fert_man_file_sys2.csv',
    'fert_man_data/fert_man_file_sys3.csv',
    'fert_man_data/fert_man_file_sys4.csv',
    'fert_man_data/fert_man_file_sys5.csv'
]

# Process each file
for file_path in file_paths:
    # Load the data
    data = pd.read_csv(file_path)
    
    # Strip leading/trailing spaces from column names
    data.columns = data.columns.str.strip()

    # Check the data types of columns
    print(f"Processing file: {file_path}")
    print(data.dtypes)
    
    # Add leading zero to date column
    data['fert date (ddmmyyyy)'] = data['fert date (ddmmyyyy)'].astype(str).apply(add_leading_zero)

    # Apply the function to the dataframe
    data = data.apply(calculate_fertilizer, axis=1)
    print(data[['fert date (ddmmyyyy)', 'fert NH4-N (g m-2).1', 'fert NO3-N banded (g m-2)', 'fert manure N (g m-2) adj']])

    # Save the updated dataframe to a new CSV file
    output_file_path = file_path.replace('.csv', '_updated.csv')
    data.to_csv(output_file_path, index=False)

    # Print output file path
    print(f"Updated file saved to: {output_file_path}")
```

```python
import csv
import os
from datetime import datetime

# Base path for input files, adjust this to your actual paths
base_input_path = '/Users/oscarli/Desktop/GitHub/eric-trial-data/fert_man_data/fert_man_file_sys{}_updated.csv'
# Base output directory, adjust to where you want to store the output folders
base_output_dir = '/Users/oscarli/Desktop/GitHub/eric-trial-data/'

# Function to parse the date and extract the year
def get_year_from_date(date_str):
    try:
        return datetime.strptime(date_str, "%d%m%Y").year
    except ValueError:
        # Handle dates with missing leading zeros
        return datetime.strptime(date_str, "%d%m%y").year

# Loop through each system
for sys_num in range(1, 6):  # From sys1 to sys5
    yearly_data = {}
    
    # Adjust file path for the current system
    file_path = base_input_path.format(sys_num)
    
    # Create a directory for the current system if it doesn't exist
    system_output_dir = os.path.join(base_output_dir, f'fert_sys_{sys_num}')
    os.makedirs(system_output_dir, exist_ok=True)
    
    # Read the CSV file for the current system
    with open(file_path, mode='r') as file:
        reader = csv.reader(file)
        headers = next(reader)  # Read the header line

        # Find the indices of the columns to be skipped
        end_index = headers.index('fertilizer type encoded') + 1
        skip_indices = {headers.index('fast or slow release fertilizer'), headers.index('fertilizer type')}

        # Adjust skip_indices to account for columns after 'fertilizer type encoded'
        skip_indices.update(range(end_index, len(headers)))
        # print(skip_indices)

        for row in reader:
            # Keep only the relevant columns
            filtered_row = [cell for i, cell in enumerate(row) if i < end_index and i not in skip_indices]
            date_str = filtered_row[0]
            year = get_year_from_date(date_str)
            if year not in yearly_data:
                yearly_data[year] = []
            yearly_data[year].append(','.join(filtered_row))
    
    # Write each year's data to a separate text file within the system-specific folder
    for year, lines in yearly_data.items():
        output_file_path = os.path.join(system_output_dir, f'fert_{year}_sys{sys_num}.txt')
        with open(output_file_path, 'w') as file:
            for line in lines:
                file.write(line + '\n')
        print(f'Data for {year} written to {output_file_path}')
```
