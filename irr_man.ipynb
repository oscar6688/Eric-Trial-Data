```python
import csv
from datetime import datetime

# Path to the CSV file
file_path = '/Users/oscarli/Desktop/GitHub/eric-trial-data/irrigation.csv'

# Function to parse the date and extract the year
def get_year_from_date(date_str):
    # Attempt to handle date formats with missing leading zeros
    try:
        return datetime.strptime(date_str, "%d%m%Y").year
    except ValueError:
        return datetime.strptime(date_str, "%d%m%y").year

# Initialize a dictionary to hold data lines for each year
yearly_data = {}

# Read the CSV file
with open(file_path, mode='r') as file:
    reader = csv.reader(file)
    next(reader)  # Skip the header line
    for row in reader:
        date_str = row[0]
        year = get_year_from_date(date_str)
        if year not in yearly_data:
            yearly_data[year] = []
        yearly_data[year].append(','.join(row))

# Write each year's data to a separate text file
for year, lines in yearly_data.items():
    output_file_path = f'/Users/oscarli/Desktop/GitHub/eric-trial-data/irr_{year}.txt'
    with open(output_file_path, 'w') as file:
        for line in lines:
            file.write(line + '\n')
    print(f'Data for {year} written to {output_file_path}')
```
